```python
def ingest_and_rotate(lines):
    all_directions = []

    # Get first direction of each row
    west_east = []
    for n in range(0, len(lines)):
        int_line = [int(x) for x in lines[n]]
        west_east.append(int_line)

    all_directions.append(west_east)

    # Get opposite direction of each row
    east_west = []
    for l in west_east:
        east_west.append([x for x in reversed(l)])
    
    all_directions.append(east_west)

    # Get rotated direction
    north_south = []
    for n in range(0, len(west_east[0])):
        vertical_line = []
        for l in west_east:
            vertical_line.append(l[n])
        north_south.append(vertical_line)

    all_directions.append(north_south)

    # Get opposite rotated direction
    south_north = []
    for l in north_south:
        south_north.append([x for x in reversed(l)])

    all_directions.append(south_north)

    return all_directions
```

```python
def line_check(line, y_val):
    coordinates = []
    # Set initial max height as first tree in line
    max_height = line[0]
    for n in range(0, len(line)-1):
        # If a tree is taller, append its coordinate and move on
        if line[n+1] > max_height:
            coordinates.append((n+1, y_val))
            max_height = line[n+1]
        # If a tree is equal or shorter, continue but do not append
        else:
            continue
    return coordinates
```

```python
def convert_coordinates(coord_pairs, orientation, max_len):
    # Return same orientation for west_east
    if orientation == 0:
        return coord_pairs

    # Reverse x coordinate for east_west
    if orientation == 1:
        coord_ret = [(max_len - x[0] - 1, x[1]) for x in coord_pairs]

    # Reverse coordinates for north_south
    if orientation == 2:
        coord_ret = [(x[1], x[0]) for x in coord_pairs]

    # Reverse y coordinates from reversed coordinates for south_north
    if orientation == 3:
        coord_ret = [(x[1], max_len - x[0] - 1) for x in coord_pairs]
        
    return coord_ret
```

```python
def eliminate_outer_coords(coord_list, max_len):
    return_list = []
    for coord in coord_list:
        if (coord[0] == max_len-1) or (coord[1] == max_len-1) or (coord[0]==0) or (coord[1]==0):
            continue
        else:
            return_list.append(coord)
    return return_list
```

```python
def check_num_of_visible_trees(grid, orientation):
    coord_pairs = []
    
    for n in range(0, len(grid)):
        line = grid[n]
        coordinates = line_check(line, n)
        coordinates = convert_coordinates(coordinates, orientation, max_len=len(line))
        coord_pairs.extend(coordinates)

    # Remove coordinate pairs which include an outer row or column
    coord_pairs = eliminate_outer_coords(coord_pairs, len(line))

    return coord_pairs
```

```python
def count_perimeter(length):
    return 4 * (length-1)
```

```python
def visible_trees(grids):
    all_trees = []
    for n in range(0, len(grids)):
        r = grids[n]
        r_trees = check_num_of_visible_trees(r, orientation=n)
        all_trees.extend(r_trees)

    all_trees = set(all_trees)
    inner_trees = len(all_trees)
    outer_trees = count_perimeter(len(grids[0]))
    return inner_trees + outer_trees
```

```python
def scenic_score(grid, coordinates):
    # Get max distance to iterate in any direction
    max_coord = len(grid[0])-1
    # Get value at coordinate
    value = grid[coordinates[1]][coordinates[0]]

    # Look north
    north_distance = 1
    for n in range(1, coordinates[1]):
        comparison_val = grid[coordinates[1]-n][coordinates[0]]
        if comparison_val < value:
            north_distance += 1
        else:
            break

    # Look south
    south_distance = 1
    for n in range(1, max_coord-coordinates[1]):
        comparison_val = grid[coordinates[1]+n][coordinates[0]]
        if comparison_val < value:
            south_distance += 1
        else:
            break

    # Look east
    east_distance = 1
    for n in range(1, max_coord-coordinates[0]):
        comparison_val = grid[coordinates[1]][coordinates[0]+n]
        if comparison_val < value:
            east_distance += 1
        else:
            break

    # Look west
    west_distance = 1
    for n in range(1, coordinates[0]):
        comparison_val = grid[coordinates[1]][coordinates[0]-n]
        if comparison_val < value:
            west_distance += 1
        else:
            break
        

    # print("NORTH DISTANCE: ", north_distance)
    # print("EAST DISTANCE: ", east_distance)
    # print("SOUTH DISTANCE: ", south_distance)
    # print("WEST DISTANCE: ", west_distance)

    return north_distance * east_distance * south_distance * west_distance
```

```python
def find_ideal_spot(grid):
    top_score = 0
    top_coord = None
    max_len = len(grid) - 1

    # Iterate over rows
    for n in range(1, max_len):
        row = grid[n]

        # Iterate over values in row
        for a in range(1, max_len):
            score = scenic_score(grid, (n,a))
            if score > top_score:
                top_score = score
                top_coord = (a, n)

    return top_score, top_coord
```
