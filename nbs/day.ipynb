```python
with open('/Users/williamkirkland/Data/KDS/adventofcode/2022/data/day1.txt') as file:
    master_list = [line.rstrip() for line in file]
```

```python
from more_itertools import split_at
```

```python
split_list = list(split_at(master_list, lambda x: x==""))
```

```python
split_list = [[int(x) for x in lst] for lst in split_list]
```

```python
split_list
```

```python
import numpy as np
```

```python
sum_list = [np.sum(x) for x in split_list]
```

```python
sum_list
```

```python
max_value = max(sum_list)
max_value
```

```python
min_value = min(sum_list)
min_value
```

```python
sorted_list = sorted(sum_list, key=lambda x: x, reverse=True)
```

```python
sorted_list
```

```python
top_3_sum = sum(sorted_list[0:3])
top_3_sum
```
